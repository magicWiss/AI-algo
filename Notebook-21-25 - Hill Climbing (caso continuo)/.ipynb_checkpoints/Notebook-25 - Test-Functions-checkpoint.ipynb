#  Funzioni di Test in Optimization


```python
import numpy as np

from numpy import asarray

from numpy import arange 
from numpy import exp 
from numpy import sqrt 
from numpy import cos 
from numpy import e
from numpy import pi

from numpy.random import randn 
from numpy.random import rand 
from numpy.random import seed

from numpy import meshgrid
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

from numpy.linalg import norm
import math

import matplotlib.pyplot as plt
```

```python
# objective function
def ackley(v):
    x, y = v
    return -20.0 * exp(-0.2 * sqrt(0.5 * (x**2 + y**2))) - exp(0.5 * (cos(2 * pi * x) + cos(2*pi*y))) + e + 20
```

## Grafico della Funzione Obiettivo

```python
# define range for input
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)
# create a mesh from the axis
x, y = meshgrid(xaxis, yaxis)
# compute targets
v = x, y
results = flower(v)
# create a surface plot with the jet color scheme
fig, ax = pyplot.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x, y, results, cmap='jet');
#pyplot.show()
```

## Test Functions

```python
# Booth's Function
def booth(v):
    x, y = v
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2    
```

```python
# Branin Function
def branin(v):
    x, y = v
    return (y - (5.1/(4*pi**2)) * x**2 + (5/pi) * x - 6)**2 + 10*(1 - (1/8*pi)) * cos(x) + 10 
```

```python
# Rosenbrock's Banana Function
def rosenbrock(v):
    x, y = v
    return (1 - x)**2 + 5*(y - x)**2   
```

```python
# Flower Function (???)
def flower(v):
    x, y = v
    return norm(v) + np.sin(4*math.atan2(x,y))
```

```python
# Wheeler's Ridge
def wheeler(v):
    x, y = v
    return -exp(-(x*y - 1.5)**2 - (y - 1.5)**2)  
```

```python
def visualizza(objective, r_min, r_max):
    # sample input range uniformly at 0.1 increments
    xaxis = arange(r_min, r_max, 0.1)
    yaxis = arange(r_min, r_max, 0.1)
    # create a mesh from the axis
    x, y = meshgrid(xaxis, yaxis)
    # compute targets
    v = x, y
    results = objective(v)
    # create a surface plot with the jet color scheme
    fig, ax = pyplot.subplots(subplot_kw={"projection": "3d"})
    ax.plot_surface(x, y, results, cmap='jet');
    #pyplot.show()
```
