#  Funzioni di Test in Optimization

I ricercatori in Optimization usano varie *test functions* per valutare algoritmi di ottimizzazione.

Qui di seguito presentiamo alcune di queste funzioni.


## Codice per visualizzazione funzioni

```python
import numpy as np
from numpy import arange 
from numpy import exp 
from numpy import sqrt 
from numpy import cos 
from numpy import e
from numpy import pi
from numpy import meshgrid
from matplotlib import pyplot
import math
import matplotlib.pyplot as plt


def visualizza(objective, r_min, r_max):
    xaxis = arange(r_min, r_max, 0.1)
    yaxis = arange(r_min, r_max, 0.1)    
    x, y = meshgrid(xaxis, yaxis) 
    v = x, y
    results = objective(v)
    fig, ax = pyplot.subplots(subplot_kw={"projection": "3d"})
    ax.plot_surface(x, y, results, cmap='jet');
    #pyplot.show()
```

## Booth's Function

E' una funzione quadratica a due variabili:

$f(x,y) = (x + 2y -7)^2 + (2x + y -5)^2$

Ha un minimo globale nel punto [1, 3] in cui ha valore zero.

```python
def booth(v):
    x, y = v
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2    
```

```python
visualizza(booth, -5, 5)
```

## Ackley Function

La funzione di Ackley è utilizzata per valutare la suscettibilità di un metodo a rimanere bloccato in minimi locali.

Tale funzione è definita per un numero qualsiasi di dimensioni d:


$f(\textbf{x}) = -a \cdot \text{exp}\left( -b \sqrt{\frac{1}{d} \sum_{i=1}^d  x_i^2} \right) - \text{exp}\left( \frac{1}{d} \sum_{i=1}^d  \cos(cx_i) \right) + a + \text{exp}(1)$

Essa ha un minimo globale all'origine in cui assume il valore zero.

Valori tipici dei parametri sono i seguenti: $a=20$, $b=0.2$, $c = 2 \pi$. Qui di seguito è presentata una implementazione per $d=2$ (due variabili).

```python
def ackley(v):
    x, y = v
    return -20.0 * exp(-0.2 * sqrt(0.5 * (x**2 + y**2))) - exp(0.5 * (cos(2 * pi * x) + cos(2*pi*y))) + e + 20
```

```python
visualizza(ackley, -5, 5)
```

## Rosenbrock's Banana Function

La Rosenbrock Function, chiamata anche Rosenbrock's Valley oppure  Rosenbrock's Banana Function, è una ben nota test function proposta da Rosenbrock nel 1960:

$f(x,y) = (a - x)^2 + b(y - x^2)^2$

Ha un minimo globale nel punto $[a, a^2]$ in cui vale zero. Nella implementazione che segue usiamo i valori $a = 1$ e $b = 5$ per i parametri.

```python
def rosenbrock(v):
    x, y = v
    return (1 - x)**2 + 5*(y - x)**2  
```

```python
visualizza(rosenbrock, -5, 5)
```

## Branin Function

Funzione a due variabili:

$f(x,y) = a(y -bx^2 + cx - r)^2 + s(1-t)\cos(x) + s$




Valori raccomandati per i parametri:

$a = 1, b = \frac{5.1}{4\pi^2},  c = \frac{5}{\pi}, r = 6, s = 10, t = \frac{1}{8\pi}$

Non ha minimi locali. Ha minimi globali in $x = \pi + 2 \pi m$ con $m$ intero, in cui la funzione vale circa $0.397887$.

```python
def branin(v):
    x, y = v
    return (y - (5.1/(4*pi**2)) * x**2 + (5/pi) * x - 6)**2 + 10*(1 - (1/8*pi)) * cos(x) + 10 
```

```python
visualizza(branin, -5, 5)
```

## Wheeler's Ridge 

Funzione a due variabili:

$f(x,y) = -exp(-(xy - a)^2 - (y - a)^2)$

Valore tipico del parametro a è 1.5, per il quale l'ottimo globale è nel punto [1, 3/2] dove la funzione vale -1.

```python
def wheeler(v):
    x, y = v
    return -exp(-(x*y - 1.5)**2 - (y - 1.5)**2)  
```

```python
visualizza(wheeler, -5, 5)
```
